```python
# Set up Dependencies
import matplotlib.pyplot as plt
from datetime import datetime
import pandas as pd
import numpy as np
import csv
import os
```

```python
# The path to our CSV file
file_1 = "data_files/Counts_of_Deaths_by_Select_Causes__2014-2019.csv"
# Read the data back into the dataframe from the csv file
data_1 = pd.read_csv(file_1)

# Drop the last 5 columns
df1 = data_1.drop(columns= ["Jurisdiction of Occurrence"])
df1.head()
```

```python
# The path to our CSV file
file_2 = "data_files/Counts_of_Deaths_by_Select_Causes__2020-2023.csv"

# Read the data back into the dataframe from the csv file
data_2 = pd.read_csv(file_2)
data_2.head()

# Drop the last 5 columns
df2 = data_2.iloc[:, 4:-5]
df2.head()
```

```python
data = pd.concat([df1, df2], ignore_index=True)
data.head()
```

```python
data["Year"] = data["Year"].astype(str)
data["Month"] = data["Month"].astype(str)
data["date"] = data["Year"] + " " + data["Month"]
data.date = data["date"].map(lambda x: datetime.strptime(x,"%Y %m"))
data["Time_Axis"] = data["date"].map(lambda x: datetime.strftime(x,"%Y-%b"))
data = data.set_index(["Time_Axis"])
data = data.sort_values(by ="date")
data = data.drop(columns=["date", "All Cause", "Drug Overdose","Assault (Homicide)", "Intentional Self-Harm (Suicide)", "Motor Vehicle Accidents", "Accidents (Unintentional Injuries)"])
file_name = "data_files/data.csv"
data.to_csv(file_name, index=False)
```
